#Scraping the Wikipedia page for the tablular data
BeautifulSoup Library will be used for web scraping of tablular data from Wikipedia page.

```python
webpage = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup=BeautifulSoup(webpage,'lxml')
print(soup.title)
from IPython.display import display_html
tab = str(soup.table)
display_html(tab,raw=True)
```

```python
# import this table into pandas data frame and display first five rows
dfwiki = pd.read_html(tab) # read html file into pandas dataframe
df=dfwiki[0]
df.head()
```

```python
# Let's now do some pre-processing and cleaning on data. 
```

```python

# We can only use the indexes where the Borough is assigned. We will first find out which rows have the Borugh as Not assigned

BorNotAssigned = df[ df['Borough'] =='Not assigned'].index

# Delete those rows from the dataframe
df.drop(BorNotAssigned , inplace=True)

```

If Borough is not assigned, then Neighborhood is also not assigned so we get rid of Neighborhood corresponding to Not Assigned vallue of Borough and print first five rows

```python
df.loc[df['Neighborhood'] =='Not assigned' , 'Neighborhood'] = df['Borough']
df.head()
```

The cells with same Borough would have same postal code but different Neighborhood. We can combine multiple rows with same postal code into a single row with its Neighborhood separated by commas. 

```python
NewDf = df.groupby(['Postal Code','Borough'], sort=False).agg( ', '.join)
```

```python
df_2=NewDf.reset_index()
df_2.head(10)
```

Now check the size of the database using shape function


```python
df_2.shape
```

Next: We need to create the geospatial coordinates of each data point. I would use http://cocl.us/Geospatial_data to read csv file to extract geographical information of each data point.

```python

!wget -q -O 'Toronto_long_lat_data.csv'  http://cocl.us/Geospatial_data
Geographical_data = pd.read_csv('Toronto_long_lat_data.csv')
Geographical_data.head()
```

```python
# Merge these columns with df_2 to make one dataframe
```

```python
Neighbor_Toronto = pd.merge(df_2,
                 Geographical_data[['Postal Code','Latitude', 'Longitude']],
                 on='Postal Code')
Neighbor_Toronto
```

Clustering the Neighborhood data. 
In this part we will first import geopy for geospatial coordinates of Toronto. 

```python
from geopy.geocoders import Nominatim # convert an address into latitude and longitude values

# Matplotlib and associated plotting modules
import matplotlib.cm as cm
import matplotlib.colors as colors

# import k-means from clustering stage
from sklearn.cluster import KMeans

#!conda install -c conda-forge folium=0.5.0 --yes # uncomment this line if you haven't completed the Foursquare API lab
import folium # map rendering library

print('Libraries imported.')
```

```python
# print geographicalinformation of City of Toronto
address = 'Toronto, ON'

geolocator = Nominatim(user_agent="Toronto")
location = geolocator.geocode(address)
latitude_toronto = location.latitude
longitude_toronto = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude_toronto, longitude_toronto))
```

Now we will use folium library to create map of the City of Toronto

```python

map_toronto = folium.Map(location=[latitude_toronto, longitude_toronto], zoom_start=10)

# add markers to map
for lat, lng, borough, Neighbourhood in zip(Neighbor_Toronto['Latitude'], Neighbor_Toronto['Longitude'], Neighbor_Toronto['Borough'], Neighbor_Toronto['Neighborhood']):
    label = '{}, {}'.format(Neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
map_toronto
```

Now we can use k-Means clustering to visualize the clustered Neighborhoods

```python
k=5
toronto_clustering = Neighbor_Toronto.drop(['Postal Code','Borough','Neighborhood'],1)
k_means = KMeans(n_clusters = k,random_state=0).fit(toronto_clustering)
k_means_labels = k_means.labels_
k_means_labels
```

```python
k_means_cluster_centers = k_means.cluster_centers_
k_means_cluster_centers
```

```python
# create map
map_clusters = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

# set color scheme for the clusters
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, neighbourhood, cluster in zip(Neighbor_Toronto['Latitude'], Neighbor_Toronto['Longitude'], Neighbor_Toronto['Neighborhood'], Neighbor_Toronto['Cluster Labels']):
    label = folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
